```python
datas['Total']=datas['math score']+datas['reading score']+datas['writing score']
```

```python
datas.head(20)
```

```python
datas.insert(9, 'Rank', datas.Total)
data1=datas[(datas.Rank>250)&(datas.Rank<300)]
data1['Rank']='Excellet'

data2=datas[(datas.Rank<250)&(datas.Rank>200)]
data2['Rank']='Very good'

data3=datas[(datas.Rank<200)&(datas.Rank>150)]
data3['Rank']='Good'

data4=datas[(datas.Rank<150)&(datas.Rank>100)]
data4['Rank']='Pass'

data5=datas[datas.Rank<=100]
data5['Rank']='Fail'
```

```python
datas=pd.concat([data1,data2,data3,data4,data5])
datas
```
